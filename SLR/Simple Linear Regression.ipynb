```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('homeprices.csv')
```

```python
df1.head()
```

```python
%matplotlib inline
plt.scatter(data = df1, x = 'area',y="price")
plt.xlabel('area')
plt.ylabel('Price')
plt.show()
```

```python
# sns.scatterplot(data = df1,x='area',y='price')
sns.relplot(data=df1,x='area',y='price',kind='scatter')
```

```python
# Creating Linear Regression Object
reg = linear_model.LinearRegression()
```

```python
#Training Linear Regression Model
#df1[['area']].values : convert into numpy array
reg.fit(df1[['area']].values,df1['price']) #First argument needs to be This makes the API consistent, 
# and your code won't need to handle special cases when you switch between single and multiple features.
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python

reg.predict([[3300]]) #needs 2-D array for input
```

```python
reg.predict([[4000]])
```

```python
%matplotlib inline
plt.scatter(data = df1, x = 'area',y="price")
plt.plot(df1['area'],reg.predict(df1[['area']].values)) #Plotting  Best Fit Line
plt.xlabel('area')
plt.ylabel('Price')
plt.show()
```
