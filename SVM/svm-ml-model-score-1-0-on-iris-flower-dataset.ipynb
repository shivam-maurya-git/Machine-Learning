## Importing Libraries

```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.svm import SVC
# They are just different implementations of the same algorithm. The SVM module (SVC, NuSVC, etc) is a wrapper
# SVC is a supervised learning algorithm that can be used for both classification and regression problems. 
# The function of SVC is to find the optimal hyperplane that can separate the given data into different classes.
```

## Loading Data

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

## Viz for Relationship among features

```python
df
```

```python
df['target'] = iris.target
```

```python
df.head()
```

```python
iris.target_names
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
```

```python
df.head()
```

```python
sns.scatterplot(data = df,x='sepal length (cm)',y='sepal width (cm)',hue='flower_name')
plt.show()
```

```python
sns.scatterplot(data = df,x='petal length (cm)',y='petal width (cm)',hue='flower_name')
plt.show()
```

## Preparing data for Training of Support Vector Machine Model

```python
x = df.drop(columns = df[['target','flower_name']])
```

```python
y = df['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

## Training Model

```python
model = SVC(C=10)
```

```python
model.fit(x_train,y_train)
```

## Model Prediction and Score

```python
model.C,model.gamma,model.kernel, model.degree
```

```python
model.score(x_test,y_test)
```
